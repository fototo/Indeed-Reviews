```python
from gensim import corpora, models # Load dictionary, corpus, model
import pyLDAvis.gensim # Generate LDA visualization
```

```python
# Load dictionary, corpus, model (positive employee reviews)
dictionary = corpora.Dictionary.load('reviews_pro.dict')
corpus = corpora.MmCorpus('reviews_pro.mm')
lda = models.LdaModel.load('lda_pro.model')
```

```python
# Generate LDA visualization
pyLDAvis.enable_notebook()
pyLDAvis.gensim.prepare(lda, corpus, dictionary)
```

```python
# Load dictionary, corpus, model (negative employee reviews)
dictionary = corpora.Dictionary.load('reviews_con.dict')
corpus = corpora.MmCorpus('reviews_con.mm')
lda = models.LdaModel.load('lda_con.model')
```

```python
# Generate LDA visualization
pyLDAvis.enable_notebook()
pyLDAvis.gensim.prepare(lda, corpus, dictionary)
```
